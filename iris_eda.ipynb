The Iris flower data set or Fisherâ€™s Iris data set is one of the most famous multivariate data set used for testing various Machine Learning Algorithms.

It includes three iris species with 50 samples each as well as some properties of each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

```python
plt.title("Species Count")
sns.countplot(df['species'], palette = "pastel")
```

```python
plt.figure(figsize = (10, 10))
plt.title("Comparision between various species based on Sepal Length and Width.")
sns.scatterplot(data= df, x='sepal_length', y='sepal_width', hue='species', s=50)
```

```python
plt.figure(figsize = (9, 9))
plt.title("Comparision between various species based on Sepal Length and Width.")
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=50)
```

```python
df.groupby('species').agg(["mean","median"])
```

```python
fig, axes = plt.subplots(2,2,figsize =(16,9))
sns.boxplot(data= df, x = 'species', y = 'petal_width', orient= 'v', ax= axes[0,0], palette='Set2')

sns.boxplot(data= df, x = 'species', y = 'petal_length', orient= 'v', ax= axes[0,1], palette='Set2')

sns.boxplot(data= df, x = 'species', y = 'sepal_width', orient= 'v', ax= axes[1,0],palette='Set2')

sns.boxplot(data= df, x = 'species', y = 'sepal_length', orient= 'v', ax= axes[1,1],palette='Set2')
plt.show()
```

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
titles = ["Sepal Length Distribution", "Sepal Width Distribution",
          "Petal Length Distribution", "Petal Width Distribution"]

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for ax, feature, title in zip(axes.flatten(), features, titles):
    g = sns.histplot(data=df, x=feature, hue="species", kde=True, ax=ax)
    ax.set_title(title)

plt.tight_layout()

plt.show()
```

```python
g = sns.FacetGrid(df, hue="species", height=5)

g.map(sns.histplot, "petal_length", kde = True)

g.add_legend()
plt.show()
```

```python
correlation_matrix = df.drop(columns='species').corr()
print(correlation_matrix)

plt.figure(figsize=(4, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

```python
sns.pairplot(data = df, hue = 'species')
plt.show()
```

Through the utilization of diverse graphical techniques like Count Plot, Scatter Plot, Box Plot, and Histogram, we conducted a comprehensive exploration of the Iris dataset, uncovering valuable insights about its features and distributions.
